```python
import pandas as pd
import matplotlib.pyplot as plt

df1.sort_values(by="Year", inplace=True)
ganga = df3['employment_type'].value_counts()
explode = [0.05 if val < ganga.max() else 0 for val in ganga.values]
g = df2.sort_values(by='jobs', ascending=False).drop_duplicates(subset='name')
b = g.head(5)
b.loc[:, 'jobs'] = b['jobs'].astype(int)
colors = ['blue','pink','red','orange','green']

fig, axs = plt.subplots(3, 2, figsize=(18, 15))
fig.suptitle("Job Trends Monitoring System", fontsize=24, fontweight='bold')

x = df1['Year'].astype(str)
y = df1['AI_Job_Percentage(%)']
axs[0, 0].bar(x, y, color='orange')
axs[0, 0].set_title("AI Job Growth Across Years", fontsize=16)
axs[0, 0].set_xlabel("Year", fontsize=14)
axs[0, 0].set_ylabel("AI Job Percentage", fontsize=12)

g = df1['Year'].astype(str)
z = df1['Layoffs']
axs[0, 1].plot(g, z, marker='o', color='r', linestyle='-', linewidth=2, markersize=6)
axs[0, 1].set_xlabel('Year', fontsize=14)
axs[0, 1].set_ylabel('No.of layoffs', fontsize=14)
axs[0, 1].set_title("Year-wise Layoffs trend", fontsize=16)

x = a['job_title']
y = a['salary']
axs[1, 0].bar(x, y, color='blue')
axs[1, 0].set_title("Top 5 Job roles", fontsize=20)
axs[1, 0].set_xlabel("Job Title", fontsize=14)
axs[1, 0].set_ylabel("salary", fontsize=12)

axs[1, 1].pie(ganga.values, labels=ganga.index, startangle=90, explode=explode, autopct='%1.1f%%')
axs[1, 1].set_title("Job Distribution BY Employee Type", fontsize=20)
axs[1, 1].legend()
axs[1, 1].axis('equal')

x = b['name']
y = b['jobs']
axs[2, 0].bar(x, y, color=colors)
axs[2, 0].set_title("Top 5 Company Hirings", fontsize=20)
axs[2, 0].set_xlabel("Company", fontsize=14)
axs[2, 0].set_ylabel("jobs", fontsize=14)
axs[2, 0].set_ylim(0, b['jobs'].max()+100)

axs[2, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
